```python
# Importation des librairies
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Chargement des donnes (Iris Dataset)
iris = load_iris()
X = iris.data
y = iris.target

# Prparation des donnes : utilisation seulement des caractristiques "petal length" et "petal width"
X = X[:, 2:4]

# Conversion des dataframes en tableau Numpy (dj fait dans ce cas)

# Affichage des donnes
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()

# chantillonnage des donnes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# S'assurer graphiquement de la bonne rpartition des donnes
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.title('Training Set')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

plt.subplot(1, 2, 2)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test)
plt.title('Test Set')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()

# Entranement du modle et valuation des performances
for depth in range(3, 11):
    clf = DecisionTreeClassifier(max_depth=depth, random_state=42)
    clf.fit(X_train, y_train)

    # valuation des performances sur le jeu d'apprentissage
    train_pred = clf.predict(X_train)
    train_accuracy = accuracy_score(y_train, train_pred)
    print(f"Max Depth: {depth}, Training Accuracy: {train_accuracy}")

    # valuation des performances sur le jeu de test
    test_pred = clf.predict(X_test)
    test_accuracy = accuracy_score(y_test, test_pred)
    print(f"Max Depth: {depth}, Test Accuracy: {test_accuracy}")
```
